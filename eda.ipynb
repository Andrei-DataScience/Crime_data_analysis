<h4>The purpose of this Exploratory Data Analysis (EDA) is to understand patterns and identify key variables, preparing th data for further analysis </h4>

<b>Importing of dependecies</b>

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

<b>Loading csv file</b>

```python
file_path = "cleaned_crime_data.csv"
crimes = pd.read_csv("cleaned_crime_data.csv")
```

```python
crimes.head()
```

<b>Getting some simple stats of numeric and time object values</b>

```python
crimes['Vict Age'].describe()
```

```python
crimes['TIME OCC'].describe()
```

```python
crimes['DATE OCC'].describe()
```

<b>Distribution of male and female victims </b>

```python
gender_counts =crimes["Vict Sex"].value_counts().head(3)
sns.barplot(data=gender_counts)
```

<b> Most common types of crime</b>

```python
crime_desc_counts =crimes["Crm Cd Desc"].value_counts().head(20)

plt.figure(figsize=(16, 12))

sns.barplot(data=crime_desc_counts)

plt.xticks(rotation=90,size=12)
plt.xlabel("Type of crime",fontsize=16)
plt.ylabel("Number of crimes",fontsize=16)

```

<b> Most common of area names </b>

```python
area_counts =crimes["AREA NAME"].value_counts()

print(area_counts)

plt.figure(figsize=(16, 12))

sns.barplot(data=area_counts)

plt.xticks(rotation=90,size=12)
plt.xlabel("Area name",fontsize=16)
plt.ylabel("Number of crimes",fontsize=16)
```
